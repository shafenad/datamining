```python
# 1

import pandas as pd
df = pd.read_csv('iris.csv')

# Definisikan fungsi normalisasi menggunakan normalisasi Min-Max
def normalize(df, column): #dataframe yang isinya data yang akan dinormalisasi dan kolom (atribut) yang akan dinormalisasi
    Vmax = df[column].max()  #menemukan nilai maxnya dari kolom
    Vmin = df[column].min() # menemukan nilai min nya
    Vbaru = (df[column] - Vmin) / (Vmax - Vmin) #variabel baru buat menghitung nilai yang ternormalisasi, rumusnya [ (nilai asli - nilai minimum) / (nilai maksimum - nilai minimum) ] --> rumus yang digunakan menskala nilai dalam kolom antara 0 dan 1
    return Vbaru

# Normalisasi setiap atribut dari iris dataframe --> menerapkan fungsi normalize
df['sepal_length'] = normalize(df, 'sepal.length')
df['sepal_width'] = normalize(df, 'sepal.width')
df['petal_length'] = normalize(df, 'petal.length')
df['petal_width'] = normalize(df, 'petal.width')

pd.DataFrame(df.head(10))
```

```python
# 2

# Menghitung Centroid Rata-rata untuk Setiap Atribut Ternormalisasi
sepal_length_centroid = df['sepal_length'].mean()
sepal_width_centroid = df['sepal_width'].mean()
petal_length_centroid = df['petal_length'].mean()
petal_width_centroid = df['petal_width'].mean()

# Print Centroid Rata-rata
print("Centroid sepal_length_norm:", sepal_length_centroid)
print("Centroid sepal_width_norm:", sepal_width_centroid)
print("Centroid petal_length_norm:", petal_length_centroid)
print("Centroid petal_width_norm:", petal_width_centroid)
```

```python
import pandas as pd
import numpy as np

# Fungsi untuk menghitung jarak Euclidean antara dua titik
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)**2))

# Load dataset iris yang telah dinormalisasi
df = pd.read_csv('iristesting.csv')

# Inisialisasi nilai centroid untuk setiap atribut
sepal_length_centroid = df['sepal.length'].mean()
sepal_width_centroid = df['sepal.width'].mean()
petal_length_centroid = df['petal.length'].mean()
petal_width_centroid = df['petal.width'].mean()

# Inisialisasi daftar kelas
classes = ['setosa', 'versicolor', 'virginica']

# Fungsi untuk memprediksi kelas data
def predict_class(data_row): #fungsi untuk memprediksi data pada setiap baris
    centroids = { #diambil dari yang udah diitung sebelumnya
        'sepal.length': sepal_length_centroid,
        'sepal.width': sepal_width_centroid,
        'petal.length': petal_length_centroid,
        'petal.width': petal_width_centroid
    }
    distances = [] #menyimpan jarak antara data dan setiap centroid kelas
    # Hitung jarak dari data ke setiap centroid kelas
    for attribute, centroid_value in centroids.items(): #iterasi melalui setiap atribut misal, sepal length dan nilai centroid yang sesuai
        distance = euclidean_distance(data_row[attribute], centroid_value) #dipanggil buat ngitung jarak euclidean antara nilai atribut pada baris data
        distances.append(distance) #dari jaraknya itu di masukkan kedalam list

# Ambil kelas dengan jarak terdekat
    min_distance_index = np.argmin(distances) #memberikan indeks dari nilai jarak terkecil, yang sesuai dengan kelas yang diprediksi
    if min_distance_index >= len(classes):  # Periksa apakah indeks valid dengan membandingkan sama panjang list classes
        return "Unknown" #kalo misal panjangnya melebihi berati di anggep kelas unknown
    else:
        predicted_class = classes[min_distance_index] #kalo kelasnya sesuai dengan indeks dari list
        return predicted_class #disimpan ke variabel

# Prediksi kelas untuk setiap baris data di iristesting.csv
df['predicted_class'] = df.apply(predict_class, axis=1) #prediksi kelas tiap data diitung pake fungsi predict class, yang dimana fungsi diterapkan disetiap baris

# Tampilkan hasil prediksi
print(df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width', 'predicted_class']])

```

```python
# Hitung jumlah kelas yang benar
predicted_class_counts = df['predicted_class'].value_counts(normalize=False)

# Periksa apakah 'setosa', 'versicolor', dan 'virginica' ada dalam hasil penghitungan
correct_class_count = 0
if 'setosa' in predicted_class_counts:
    correct_class_count += predicted_class_counts['setosa']
if 'versicolor' in predicted_class_counts:
    correct_class_count += predicted_class_counts['versicolor']
if 'virginica' in predicted_class_counts:
    correct_class_count += predicted_class_counts['virginica']

# Hitung jumlah total kelas yang diprediksi
total_class_count = predicted_class_counts.sum()

# Hitung persentase klasifikasi yang benar
accuracy = (correct_class_count / total_class_count) * 100

print(f"Accuracy: {accuracy}%")
```
